###Text Summarization & Sentiment Analysis

```python
import pandas as pd
import numpy as np
import spacy
import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk import sent_tokenize
nltk.download('punkt')
nltk.download('stopwords')
import heapq
```

```python
data = open('/content/sample_data/spaceX_DP.txt').read()
data
```

```python
#Removing stopwords and punctuations
clean_data = re.sub(r'[^\w\s]', ' ', data)
clean_data
clean_data=clean_data.lower()
clean_data
```

```python
#Tokenize the 'SpaceX.txt' corpus.
tokens = nltk.word_tokenize(clean_data)
print(tokens)
```

```python
#Determine the maximum frequency as 'word_frequencies[word]=(word_frequencies[word]/maximum_frequency). 
stop_words = nltk.corpus.stopwords.words('english')
word_frequencies = {}
for word in nltk.word_tokenize(clean_data):
    if word not in stop_words:
        if word not in word_frequencies.keys():
            word_frequencies[word] = 1
        else:
            word_frequencies[word] += 1


maximum_frequncy = max(word_frequencies.values())

for word in word_frequencies.keys():
    word_frequencies[word] = (word_frequencies[word]/maximum_frequncy)


print(word_frequencies)
```

```python
#Tokenize the sentences. Generate the sentence_scores. Score every sentence based on number of words. 
clean_data1 = re.sub(r'\[[0-9]*\]', ' ', data)
clean_data1 = re.sub(r'\s+', ' ', data)
sen_list = nltk.sent_tokenize(clean_data1)
sen_list
```

```python
sen_scores = {}
for sentence in sen_list:
    for word in nltk.word_tokenize(sentence.lower()):
        if word in word_frequencies.keys():
            if len(sentence.split(' ')) < 30:
                if sentence not in sen_scores.keys():
                    sen_scores[sentence] = word_frequencies[word]
                else:
                    sen_scores[sentence] += word_frequencies[word]

sen_scores
```

```python
#Import nlargest from heapq and provide 'summarized_sentences'.
summarized_sentences = heapq.nlargest(10,sen_scores, key=sen_scores.get) 
summarized_sentences
```

```python
#Convert sentences from spacy to strings and join all sentences.
words=[word for word in summarized_sentences]
words
final_summary =''.join(words)
final_summary
```

```python
# Determine the length of the summary & determine the length of the original text
print("Summary Length:",len(final_summary))
print("original text Length:",len(data))
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
```

```python
# Loading the file
data = open('/content/sample_data/SpaceX.txt').read()
data
```

```python
def sentiment_scores(sentence):
 
    # Create a SentimentIntensityAnalyzer object.
    sid_obj = SentimentIntensityAnalyzer()
 
    # polarity_scores method of SentimentIntensityAnalyzer
    # object gives a sentiment dictionary.
    # which contains pos, neg, neu, and compound scores.
    sentiment_dict = sid_obj.polarity_scores(sentence)
     
    print("Overall sentiment dictionary is : ", sentiment_dict)
    print("sentence was rated as ", sentiment_dict['neg']*100, "% Negative")
    print("sentence was rated as ", sentiment_dict['neu']*100, "% Neutral")
    print("sentence was rated as ", sentiment_dict['pos']*100, "% Positive")
 
    print("Sentence Overall Rated As", end = " ")
 
    # decide sentiment as positive, negative and neutral
    if sentiment_dict['compound'] >= 0.05 :
        print("Positive")
 
    elif sentiment_dict['compound'] <= - 0.05 :
        print("Negative")
 
    else :
        print("Neutral")
 
 
   
# Driver code
if __name__ == "__main__" :
  sentiment_scores(data)
```

### Is the text more ‘subjective’ or more ‘objective’?
The overall sentiment is objective since the text's sentiment is neutral and the author is not attempting to sway the readers' opinions. We could have stated that the entire text is subjective if the text's conclusion had been positive or negative. The text is therefore very objective and neutral.
